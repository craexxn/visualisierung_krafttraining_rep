```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the dummy data from the CSV file
dummy_data = pd.read_csv('data/dummy_training_data.csv')

# Convert the date column to datetime format
dummy_data['date'] = pd.to_datetime(dummy_data['date'])

# Convert the date column to datetime format
dummy_data['date'] = pd.to_datetime(dummy_data['date'])

# Extract month and year from the date column
dummy_data['year_month'] = dummy_data['date'].dt.to_period('M')

# Calculate the total weight lifted per exercise per month
total_weight_per_exercise_monthly = dummy_data.groupby(['year_month', 'exercise'])['total_weight'].sum().reset_index()

# Plot the data
plt.figure(figsize=(14, 7))
sns.barplot(x='year_month', y='total_weight', hue='exercise', data=total_weight_per_exercise_monthly)
plt.title('Total Weight Lifted per Exercise (Monthly)')
plt.xlabel('Month')
plt.ylabel('Total Weight (kg)')
plt.xticks(rotation=45)
plt.legend(title='Exercise')
plt.show()
```
