```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the dummy data from the CSV file
dummy_data = pd.read_csv('data/dummy_training_data.csv')

# Calculate the estimated 1RM using the Epley formula
dummy_data['estimated_1RM'] = dummy_data['weight'] * (1 + 0.0333 * dummy_data['reps'])

# Calculate the average estimated 1RM per day per exercise
average_1RM_per_day = dummy_data.groupby(['date', 'exercise'])['estimated_1RM'].mean().unstack()

# Apply a moving average to smooth the lines (window size of 7 days)
smoothed_1RM_per_day = average_1RM_per_day.rolling(window=7, min_periods=1).mean()

# Plot the data
plt.figure(figsize=(14, 7))
for exercise in smoothed_1RM_per_day.columns:
    plt.plot(smoothed_1RM_per_day.index, smoothed_1RM_per_day[exercise], label=exercise)

plt.title('Smoothed Estimated One-Rep Max (1RM) Over Time')
plt.xlabel('Date')
plt.ylabel('Estimated 1RM (kg)')
plt.legend(title='Exercise')
plt.grid(True)
plt.show()
```
