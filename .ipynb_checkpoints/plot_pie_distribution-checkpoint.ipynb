```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the dummy data from the CSV file
dummy_data = pd.read_csv('data/dummy_training_data.csv')

# Initial max values for each exercise at the beginning of the year
initial_max_values = {
    'squat': 150,
    'deadlift': 200,
    'bench press': 135
}

# Calculate the percentage of max weight lifted for each exercise
dummy_data['percentage_of_max'] = dummy_data.apply(
    lambda row: (row['weight'] / initial_max_values[row['exercise']]) * 100, axis=1)

# Define categories based on percentage of max weight
bins = [0, 50, 70, 85, 100]
labels = ['0-50%', '50-70%', '70-85%', '85-100%']

# Create a column for these categories
dummy_data['percentage_category'] = pd.cut(dummy_data['percentage_of_max'], bins=bins, labels=labels)

# Plot the pie charts for each exercise
exercises = ['squat', 'bench press', 'deadlift']
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

fig, axs = plt.subplots(1, 3, figsize=(18, 6))

for i, exercise in enumerate(exercises):
    exercise_data = dummy_data[dummy_data['exercise'] == exercise]
    category_counts = exercise_data['percentage_category'].value_counts().sort_index()
    
    axs[i].pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=140, colors=colors)
    axs[i].set_title(f'{exercise.capitalize()} Training Intensity Distribution')

plt.tight_layout()
plt.show()
```
