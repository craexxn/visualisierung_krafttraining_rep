```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the dummy data from the CSV file
dummy_data = pd.read_csv('data/dummy_training_data.csv')

# Calculate the average weight lifted per day per exercise
average_weight_per_day = dummy_data.groupby(['date', 'exercise'])['weight'].mean().unstack()

# Plot the data
average_weight_per_day.plot(kind='line', figsize=(14, 7))
plt.title('Average Weight Lifted Over Time')
plt.xlabel('Date')
plt.ylabel('Average Weight (kg)')
plt.legend(title='Exercise')
plt.grid(True)
plt.show()
```
