```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the dummy data from the CSV file
dummy_data = pd.read_csv('data/dummy_training_data.csv')

# Convert the date column to datetime format
dummy_data['date'] = pd.to_datetime(dummy_data['date'])

# Extract month and year from the date column
dummy_data['year_month'] = dummy_data['date'].dt.to_period('M').astype(str)

# Calculate the training frequency per month per exercise
training_frequency_per_month = dummy_data.groupby(['year_month', 'exercise']).size().reset_index(name='count')

# Plot the data
plt.figure(figsize=(14, 7))
sns.scatterplot(data=training_frequency_per_month, x='year_month', y='count', hue='exercise', s=100, alpha=0.7)
plt.title('Training Frequency per Month by Exercise')
plt.xlabel('Month')
plt.ylabel('Number of Training Sessions')
plt.xticks(rotation=45)
plt.legend(title='Exercise')
plt.grid(True)
plt.show()
```
